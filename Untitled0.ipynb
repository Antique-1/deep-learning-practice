#데이터 불러오기

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
breast_cancer_data = datasets.load_breast_cancer()
df_2 = pd.DataFrame(data=breast_cancer_data.data, columns=breast_cancer_data.feature_names)
df_2['target'] = breast_cancer_data.target
df_2
```

#데이터 분석

```python
df_2.head(5)
```

```python
df_2.describe()
```

```python
df_2.corr()
```

```python
colormap = plt.cm.gist_heat   # 그래프 색상 구성
plt.figure(figsize=(25,25))   # 그래프 크기
```

```python
sns.heatmap(df_2.corr(),linewidths=0.1,vmax=0.5, cmap=colormap, linecolor='white', annot=False) # 그래프 시각적 구성요소를 설정하는 코드
plt.show()
```

```python
df_2.hist(bins=20, figsize=(15, 15))
plt.show()
```
